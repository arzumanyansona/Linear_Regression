```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("/content/realest.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
target = data["Price"]
target = target.drop(labels=[156], axis = 0)
```

```python
data= data.drop(["Price"], axis = 1)
```

```python
data
```

```python
data=data.drop(labels=[156], axis =0)
```

```python
data
```

```python
data.isnull().sum()

```

```python
data["Space"]=data["Space"].fillna(data['Space'].mean())
```

```python
data["Lot"]=data["Lot"].fillna(data['Lot'].mean())
data["Tax"]=data["Tax"].fillna(data['Tax'].mean())
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test,y_train, y_test= train_test_split(data, target, test_size =0.2, random_state = 42)
```

```python
X_train.shape, X_test.shape
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(criterion='squared_error', splitter='best', max_depth=10, min_samples_leaf=4, max_leaf_nodes= 100, random_state = 1)
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
pred_train = regressor.predict(X_train)
```

```python

from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test, y_pred), mean_squared_error(y_train, pred_train)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
